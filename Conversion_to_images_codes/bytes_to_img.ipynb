```python
images_path_Gatak = '/home/sanjeev/ML_Dataset/microsoft_dataset/bytes_img/Gatak/'
```

```python
img_list_1 = []
for img in os.listdir(Gatak):
    Gatak_path = os.path.join(Gatak, img)
    img_list_1.append(Gatak_path)
```

```python
for j in img_list_1:
    f = open(j, 'rb')
    ln = os.path.getsize(j)

    if ln < 10240:
        width = 32
    elif ln < 30720:
        width = 64
    elif ln < 61440:
        width = 128
    elif ln < 102400:
        width = 256
    elif ln < 204800:
        width = 384
    elif ln < 1024000:
        width = 512
    elif ln < 1536000:
        width = 1024
    else:
        width = 2048

    rem = ln%width
    a = array.array("B")
    a.fromfile(f, ln-rem)
    f.close()
    g = np.reshape(a, (len(a)//width, width))
    g = np.uint8(g)
    
    #scipy.misc.imsave('/home/ubuntu/Desktop/BinImg/%s.png'%j, g)
    #print("Image saved for bin_id %s"%j)
    
    img = Image.fromarray(g)
    print(img)
   # str1 = Gatak_img + j
    
    #result = hashlib.md5(str1.encode())
    #str2 = result.hexdigest()
    img_asm = j.split('/')[-1]
    img_name = img_asm.split('.')[0] +'.png'
    img.save((images_path_Gatak+'/')+ img_name)
```
